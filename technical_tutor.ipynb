## ðŸ“„ Digital Brochure Generator

An automated solution designed to generate high-impact corporate brochures from minimal data. Simply provide the company name and website, and the tool creates informative material tailored to different strategic audiences.

#### ðŸš€ Project Description
This tool streamlines the creation of marketing and sales materials. Using the company website as the primary source, it extracts, synthesizes, and structures key information to generate customized brochures for three main profiles:
* **Potential Clients:** Focus on products, services, and value proposition.

* **Investors:** Focus on metrics, vision, market, and business stability.

* **Potential Candidates (Recruiting):** Focus on culture, benefits, and mission.

#### ðŸ›  Inputs and Outputs
* **Input:** Company name and website URL.
* **Output:** Structured digital brochure (PDF/Markdown).

````python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
````

##### Define la personalidad del Tutor

````python
system_prompt = """
        You are an expert Technical Tutor specializing in Software Engineering and AI.
        Your goal is not just to give answers, but to help the user understand the 'Why' and 'How'.
        
        Guidelines:
        1. Explain concepts clearly using analogies.
        2. Provide code examples in Python when relevant.
        3. If the user asks a vague question, ask clarifying questions.
        4. Use professional formatting (Markdown, bolding for key terms).
        """
````

````python
load_dotenv(override=True)

class TechnicalTutor:
    def __init__(self, provider="openai", model_name="gpt-4o-mini"):
        self.provider = provider.lower()
        self.model_name = model_name

        # Customer configuration according to the provider
        if self.provider == "openai":
            api_key = os.getenv('OPENAI_API_KEY')
            self.client = OpenAI(api_key=api_key)
            print(f"ðŸŸ¢ Tutor initialized with OpenAI ({self.model_name})")
            
        elif self.provider == "ollama":
            self.client = OpenAI(
                base_url="http://localhost:11434/v1",
                api_key="ollama"
            )
            print(f"ðŸŸ  Tutor initialized with Ollama Local ({self.model_name})")
        else:
            raise ValueError("Provider must be 'openai' or 'ollama'")

    def ask(self, question):
        """
        Send the question to the configured model and display the answer in streaming.
        """
        print(f"\nThinking with {self.provider.upper()}...\n")
        
        try:
            stream = self.client.chat.completions.create(
                model=self.model_name,
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": question}
                ],
                stream=True
            )
            
            response_text = ""
            display_handle = display(Markdown("Wait..."), display_id=True)
            
            for chunk in stream:
                if chunk.choices[0].delta.content:
                    response_text += chunk.choices[0].delta.content
                    response_text = response_text.replace("```","").replace("markdown", "")
                    update_display(Markdown(response_text), display_id=display_handle.display_id)
            
            return response_text

        except Exception as e:
            print(f"Error connecting to {self.provider}: {e}")
            if self.provider == "ollama":
                print("Tip: Make sure Ollama is running explicitly (e.g., 'ollama serve' in terminal).")
````

##### EXAMPLE OF USE

````python
# 1. Instance using GPT-4o-mini (Cloud)
# Ideal for complex reasoning or very up-to-date general knowledge questions.
cloud_tutor = TechnicalTutor(provider="openai", model_name="gpt-4o-mini")
cloud_tutor.ask("Explain the difference between lists and tuples in Python using an analogy.");
````

````python
# 2. Instance using Ollama (Local)
# Ideal for privacy, zero costs, or repetitive queries.

# Make sure you have the model downloaded (e.g., `ollama pull llama3.2`)
local_tutor = TechnicalTutor(provider="ollama", model_name="llama3.2") 
local_tutor.ask("Â¿CÃ³mo funciona el garbage collector en Java?")
````
